```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2


df = pd.read_csv("uber.csv")  


def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers

    
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c


df['trip_distance'] = df.apply(lambda row: haversine_distance(
    row['pickup_latitude'],
    row['pickup_longitude'],
    row['dropoff_latitude'],
    row['dropoff_longitude']), axis=1)


df = df[df['trip_distance'] > 0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='trip_distance', y='fare_amount')
plt.title('Fare Amount vs. Distance Traveled')
plt.xlabel('Distance (km)')
plt.ylabel('Fare Amount ($)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Make sure the pickup datetime is in datetime format
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

# Extract new time-based features
df['hour'] = df['pickup_datetime'].dt.hour
df['day'] = df['pickup_datetime'].dt.day
df['month'] = df['pickup_datetime'].dt.month
```

```python
# Monday=0, Sunday=6
df['day_of_week'] = df['pickup_datetime'].dt.dayofweek

# Optional (to make it more readable)
day_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
           4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['day_name'] = df['day_of_week'].map(day_map)
```

```python
def label_peak(hour):
    if 7 <= hour <= 9 or 17 <= hour <= 19:
        return 'Peak'
    else:
        return 'Off-Peak'

df['period'] = df['hour'].apply(label_peak)
```

```python
# Ensure correct data types
df['day_name'] = df['day_name'].astype('category')
df['period'] = df['period'].astype('category')
```

```python
df.to_csv("enhanced_uber_fare.csv", index=False)
```
